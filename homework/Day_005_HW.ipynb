### 欄位的平均數

```python
print("AMT_INCOME_TOTAL 的平均數:",df['AMT_INCOME_TOTAL'].mean())
print("AMT_CREDIT 的平均數:",df['AMT_CREDIT'].mean())
print("AMT_ANNUITY 的平均數:",df['AMT_ANNUITY'].mean())
```

### 欄位的標準差

```python
print("AMT_INCOME_TOTAL 的標準差:",df['AMT_INCOME_TOTAL'].std())
print("AMT_CREDIT 的標準差:",df['AMT_CREDIT'].std())
print("AMT_ANNUITY 的標準差:",df['AMT_ANNUITY'].std())
```

```python
INCOME = app_train['AMT_INCOME_TOTAL'].mean()
CREDIT = app_train['AMT_CREDIT'].mean()
ANNUITY = app_train['AMT_ANNUITY'].mean()
```
